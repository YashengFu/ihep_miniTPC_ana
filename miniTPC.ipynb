```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
```

```python
power = [234,234,216,216,216,198,198]
voltage = [1.80,1.80,1.80,1.80,1.80,1.80,1.80]
current = [0.13,0.13,0.12,0.12,0.12,0.11,0.11]
temperature = [0,-20,-40,-60,-80,-100,-110]

plt.figure(figsize=(6,4))
plt.tick_params(labelsize=13)
x_major_locator=MultipleLocator(10)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)

for xy in zip(temperature, power):
    if xy == (-110,198):
        plt.annotate("(%s,%s)" % xy, xy=xy, xytext=(-15, 30), textcoords='offset points')
    else:
        plt.annotate("(%s,%s)" % xy, xy=xy, xytext=(-25, 10), textcoords='offset points')
plt.plot(temperature,power,'o',color='fuchsia',label='V21 charge tile')
plt.legend(prop = {'size':13})
plt.xlabel('Temperature($^\circ$C)',size=15,labelpad=10)
plt.ylabel('Power(mW)',size=15,labelpad=10)
plt.title('The change of charge tile power with temperature',size=15,pad=25)
plt.ylim(195,240)
```
