```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
noise_V21 = [622.54,628.28,590.98,588.11,550.82,539.34,485.79,465.71]
noise_V20 = [515.89,504.38,515.89,535.07,452.60,403.69,435.34,393.15]
temperature = [20,0,-20,-40,-60,-80,-100,-110]

plt.figure(figsize=(6,4))
plt.tick_params(labelsize=13)

plt.plot(temperature,noise_V21,'>',color='darkorange',label='V21 charge tile')
plt.plot(temperature,noise_V20,'b<',label='V20 charge tile')
plt.legend(prop = {'size':13})
plt.xlabel('Temperature($^\circ$C)',size=15,labelpad=10)
plt.ylabel('Noise (e-)',size=15,labelpad=10)
plt.title('The change of charge tile noise with temperature',size=15,pad=25)
```
